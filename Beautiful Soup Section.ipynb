```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://www.bls.gov/emp/tables/employment-by-major-industry-sector.htm"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
print(soup.text)
```

```python
def extract_source(url):
    agent = {"User-Agent": "Mozilla/5.0"}
    source = requests.get(url, headers=agent,verify=False).text
    return source
def extract_data(source):
    soup= BeautifulSoup(source, 'html.parser')
    return(soup)
extract_data(extract_source('https://bls.gov/emp/tables/employment-by-major-industry-sector.htm'))
```

```python
snowy =extract_data(extract_source('https://bls.gov/emp/tables/employment-by-major-industry-sector.htm'))
```
